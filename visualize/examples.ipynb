### Visualize Field Dump Example

````python
import matplotlib.pyplot as plt
from vizfielddump import VizFieldDump

file_name = 'your path to fieldump.nc'

field = VizFieldDump(file_name)
````

````python
fig,ax = field.quiver_plot('u','v','xy',stride=3,timestamp=10,x3_idx=5,background_field='ekh')
plt.show()
````

### Visualize Concentration Results

supports the output format from https://github.com/lorenzodonadio/adepostles

which is ```dimensions:
        time 
        x 
        y 
        z ```

````python
from vizconcentration import VizConcentration
import numpy as np

file_name = 'your path to concentration output netcdf file'
ibm_file = 'path to ibm.inp.001'

ibm_rows = 256

with open(ibm_file) as f:
    lines = [float(l.strip().replace("\n", "")) for l in f.readlines() if not l.startswith("#")]
    bc_height = np.flipud(np.array(lines).reshape(-1, ibm_rows))

conc = VizConcentration(file_name, bc_height)
````

````python
#try also fig,ax = conc.plot_interactive_slice()
fig,ax = conc.plot_slice(time_index=2,zidx=4,nsv=1)
plt.show()
````
